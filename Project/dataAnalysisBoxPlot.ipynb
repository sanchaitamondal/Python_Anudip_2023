```python
dataAnalysis=pd.read_excel("dataAnalysis.xlsx", sheet_name='DifferentFormat')
dataAnalysisDf=pd.DataFrame(dataAnalysis)
dataAnalysisDf.boxplot(column=['TPT'],by=['Team Member'])
plt.title("Boxplot by Team Members")
plt.xticks(rotation=90)
plt.xlabel("Team Members")
plt.ylabel("TPT")
plt.show()
```
